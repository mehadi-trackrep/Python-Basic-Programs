```python
import numpy as np
a = np.array([(1,2,3,),(5,6,7)])
b = np.array([(1,2,3,),(5,6,7)])

print(a+b) ## element wise operation hobe..
vstack = np.vstack( (a,b) )
print(vstack)

hstack = np.hstack( (a,b) )
print(hstack)
```
