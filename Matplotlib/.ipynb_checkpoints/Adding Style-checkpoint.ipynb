```python
import matplotlib.pyplot as plt
from matplotlib import style 
style.use('ggplot')

X = [1,2,3]
Y = [3,6,1]

X1 = [3,6,1]
Y1 = [1,2,3]

plt.plot(X,Y, color='g',label='first_graph', linewidth=5)
plt.plot(X1,Y1, color='c',label='second_graph', linewidth=5)

plt.title('info')
plt.xlabel('X-label')
plt.ylabel('Y-label')

plt.legend()
plt.grid(True, color='K')
plt.show()
```

```python
###Bar plot:
plt.bar(X,Y,label='Example one', color='b')
plt.bar(X1,Y1,label='Example one', color='g')

plt.legend()
plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('info')

plt.grid(True, color='K')

plt.show()
```

```python
###Histogram
import numpy as np
population_ages = [10,20,30,40,20,50,60,7,80,97,90,45,64,32]
bins = [10,20,30,40,50,62,70,80]

plt.hist(population_ages, bins, histtype='bar', rwidth=0.8)
# plt.hist(np.random.exponential(size=10000),bins=100,histtype='bar')

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Histogram')

plt.show()
```

```python
###Scatter graph

X = [1,2,3,4,5,6,7,8]
Y = [8,7,1,5,3,4,2,6]


plt.scatter(X,Y,label='first_graph', linewidth=5, color='g')

plt.title('Scatter plot')
plt.xlabel('X-label')
plt.ylabel('Y-label')

plt.legend()
plt.grid(True, color='K')
plt.show()
```

```python
###Stack/Area plot

days = [1,2,3,4,5]

sleeping = [8,7,1,5,3]
eating = [3,4,5,2,1]
working = [6,7,2,1,4]
playing = [3,4,6,5,7]


plt.plot([],[], label='sleeping', linewidth=5, color='g')
plt.plot([],[], label='eating', linewidth=5, color='b')
plt.plot([],[], label='working', linewidth=5, color='m')
plt.plot([],[], label='playing', linewidth=5, color='k')

plt.stackplot(days, sleeping,eating,working,playing, colors=['r','g','b','m','k'])

plt.title('Area plot')
plt.xlabel('X-label')
plt.ylabel('Y-label')

plt.legend()
plt.grid(True, color='K')
plt.show()
```
